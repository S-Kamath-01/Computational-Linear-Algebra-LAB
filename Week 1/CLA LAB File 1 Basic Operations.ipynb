```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotate_point(point, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians)],
                                [np.sin(angle_radians), np.cos(angle_radians)]])
    return np.dot(rotation_matrix, point)


# Example usage
point = np.array([1, 0])
theta = 60  # Enter the angle in degrees

rotated_point = rotate_point(point, theta)
print("Original Point:", point)
print("Rotated Point:", rotated_point)


# Plot the original and rotated points as vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, point[0], point[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, rotated_point[0], rotated_point[1], angles='xy', scale_units='xy', scale=1, color='red', label='Rotated Vector')

# Set limits and labels
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Rotated Vectors')

plt.show()
```

```python
def reflect_point(point, line_angle_degrees):
    line_angle_radians = np.radians(line_angle_degrees)
    reflection_matrix = np.array([
        [np.cos(2 * line_angle_radians), np.sin(2 * line_angle_radians)],
        [np.sin(2 * line_angle_radians), -np.cos(2 * line_angle_radians)]
    ])
    return np.dot(reflection_matrix, point)


# Example usage
point = np.array([1, 0])
theta = 90  # Enter the angle made by the reflection axis with positive x-axis (in degrees)

reflected_point = reflect_point(point, theta)
print("Original Point:", point)
print("Reflected Point:", reflected_point)


# Plot the original and rotated points as vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, point[0], point[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, reflected_point[0], reflected_point[1], angles='xy', scale_units='xy', scale=1, color='red', label='Reflected Vector')

# Set limits and labels
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Reflected Vectors')

plt.show()
```

```python
def scale_point(point, scale_factor):
    scaling_matrix = np.array([
        [scale_factor, 0],
        [0, scale_factor]
    ])
    return np.dot(scaling_matrix, point)


# Example usage
point = np.array([1, 0])
scale_factor = 2.5   # Enter the scaling factor

scaled_point = scale_point(point, scale_factor)
print("Original Point:", point)
print("Scaled Point:", scaled_point)


# Plot the original and rotated points as vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, point[0], point[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, scaled_point[0], scaled_point[1], angles='xy', scale_units='xy', scale=1, color='red', label='Scaled Vector')

# Set limits and labels
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Scaled Vectors')

plt.show()
```

```python
# Function to plot a shape with vertex labels
    
def plot_shape(vertices, color='blue', label_prefix='P'):
    """
    Plots a shape given by its vertices and labels each vertex.
    
    Parameters:
    - vertices: A numpy array of shape (n, 2) where n is the number of vertices.
    - color: The color of the shape and the labels.
    - label_prefix: A prefix for labeling vertices (e.g., 'P' for P1, P2, ...).
    """
    # Close the shape by appending the first vertex at the end
    closed_vertices = np.append(vertices, [vertices[0]], axis=0)
    
    # Plot the shape
    plt.plot(closed_vertices[:, 0], closed_vertices[:, 1], 'o-', color=color)
    
    # Label the vertices
    for i, (x, y) in enumerate(vertices):
        plt.text(x + 0.1, y + 0.1, f'{label_prefix}{i+1}', fontsize=12, color=color)
    
 
```

```python
# Define vertices of a square
square = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])
```

```python
# Rotate the square
theta = 60   # Angle in degrees
rotated_square = np.array([rotate_point(point, theta) for point in square])


# Plot original square
plot_shape(square, color='blue', label_prefix='P')

# Plot rotated square
plot_shape(rotated_square, color='red', label_prefix="P'")

# Set plot limits and configuration

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend(['Original Square', 'Rotated Square'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Rotated Squares')

plt.show()
```

```python
# Reflect the square
theta = 90    # Angle in degrees
reflected_square = np.array([reflect_point(point, theta) for point in square])


# Plot original square
plot_shape(square, color='blue', label_prefix='P')

# Plot reflected square
plot_shape(reflected_square, color='red', label_prefix="P'")

# Set plot limits and configuration

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend(['Original Square', 'Reflected Square'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Reflected Squares')

plt.show()
```

```python
# Scale the square
scale_factor = 2.1    # Scaling_factor
scaled_square = np.array([scale_point(point, scale_factor) for point in square])


# Plot original square
plot_shape(square, color='blue', label_prefix='P')

# Plot scaled square
plot_shape(scaled_square, color='red', label_prefix="P'")

# Set plot limits and configuration

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend(['Original Square', 'Scaled Square'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Scaled Squares')

plt.show()
```

```python
def shear_x(point, k_x):
    shear_matrix_x = np.array([[1, k_x], [0, 1]])
    return np.dot(shear_matrix_x, point)

def shear_y(point, k_y):
    shear_matrix_y = np.array([[1, 0], [k_y, 1]])
    return np.dot(shear_matrix_y, point)


# Example usage
point = np.array([1, 1])
k_x = 2  # Shear factor along x-axis
k_y = 3  # Shear factor along y-axis

sheared_point_x = shear_x(point, k_x)
sheared_point_y = shear_y(point, k_y)

print("Original Point:", point)
print("Sheared Point along X-axis:", sheared_point_x)
print("Sheared Point along Y-axis:", sheared_point_y)


# Plot the original and sheared points as vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, point[0], point[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, sheared_point_x[0], sheared_point_x[1], angles='xy', scale_units='xy', scale=1, color='red', 
           label='X-Sheared Vector')
plt.quiver(0, 0, sheared_point_y[0], sheared_point_y[1], angles='xy', scale_units='xy', scale=1, color='magenta', 
           label='Y-Sheared Vector')

# Set limits and labels
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Sheared Vectors')

plt.show()
```

```python
# Define vertices of a pentagon
pentagon = np.array([[0, 1], [0.951, 0.309], [0.588, -0.809], [-0.588, -0.809], [-0.951, 0.309]])

# Apply shear transformations
k_x = 0.5  # Shear factor along x-axis
k_y = 0.3  # Shear factor along y-axis

sheared_pentagon_x = np.array([shear_x(point, k_x) for point in pentagon])
sheared_pentagon_y = np.array([shear_y(point, k_y) for point in pentagon])

# Plot the original and sheared pentagons
plot_shape(pentagon, color='b', label_prefix='P')
plot_shape(sheared_pentagon_x, color='g', label_prefix='P')

# Set plot limits and configuration

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend(['Original Pentagon', 'Sheared Pentagon (X-axis)'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and X-Sheared Pentagons')

plt.show()
```

```python
# Plot the original and sheared pentagons
plot_shape(pentagon, color='b', label_prefix='P')
plot_shape(sheared_pentagon_y, color='r', label_prefix='P')

# Set plot limits and configuration

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend(['Original Pentagon', 'Sheared Pentagon (Y-axis)'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original and Y-Sheared Pentagons')

plt.show()
```
